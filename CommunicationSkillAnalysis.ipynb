### Provide a keyword and get the most prevalent communication type

### Provide a keyword and get communication phrases

```python
file = "merged_data_final.csv"
df = pd.read_csv("data/" + file)
keyword = "Computer Hardware Engineer"
df = df[df["Keyword"] == keyword]
df.head()
```

```python
# Update to loop over all descriptions for this job field
exDescription = df["Description"].to_numpy()[0].lower()
exDescription
```

```python
file = "coded_list_items.csv"
df2 = pd.read_csv("data/" + file)
df2.head()
```

```python
def detectSkills(phrase, skillFields):
    skills = [i for i in skillFields if isinstance(i, str)]
    
    # ADD MORE CASES IF NEEDED
    if 'write' in phrase or 'written' in phrase:
        skills += ['written']
    if 'verbal' in phrase:
        skills += ['verbal']
    if 'aural' in phrase:
        skills += ['aural']
    
    return list(set(skills))
```

```python
allSkills = []
for row in df2.to_numpy():
    phrase = row[2].lower()
    if phrase in exDescription:
        skillFields = np.array(row[-4:])
        skills = detectSkills(phrase, skillFields)
        allSkills += skills
        print(phrase)
        print(skills)
        print('---')
print(allSkills)
```

```python
y = dict((val, allSkills.count(val)) for val in set(allSkills))
```
